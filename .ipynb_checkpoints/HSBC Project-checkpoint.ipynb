```python
class DataFrame_Preprocessing:
    def __init__(self,df,if_time_series,label_col=None):
        self.data = df
        self.label_col = label_col
        self.text_col = None
        self.cat_col = None
        self.num_col = None
        self.string_cat_col = None

    def __getitem__(self, index):
        return self.data.iloc[index:index+1,:]
    def __len__(self):
        return len(self.data) 
       
    def update(self):
        self.columns = self.data.columns
        data_type=self.data_type
        self.cat_col = [i for i in data_type.loc[data_type.loc[:,'Type']== 'categorical','Column'] if i != self.label_col]
        self.num_col = [i for i in data_type.loc[data_type.loc[:,'Type']== 'numerical','Column'] if i != self.label_col]    
        self.string_cat_col = [i for i in self.columns if (str(self.data.loc[:,i].dtypes) in ['object','bool'] and i != self.label_col)]
        self.training_data = list(set(self.training_data) & set(self.data.index))  # remove entries that no longer exist
        self.testing_data = list(set(self.testing_data) & set(self.data.index))
        return None 
    
    def Convert_numerical(self,col_names,noise = ','):
        # deal with wrong format number 
        # comma as parameter !!!
        for i in col_names:
            self.data.loc[:,i] = self.data.loc[:,i].apply(lambda x: float(str(x).replace(' ','').replace(noise,'')))
        return None
    
    def set_sequential_string_columns(self,column=None):
        self.text_col = column
        return None        
    
    @property
    def data_quality_score(self):
        import warnings
        warnings.filterwarnings("ignore")
        from sklearn.preprocessing import LabelEncoder
        from sklearn.impute import SimpleImputer
        from sklearn.metrics import average_precision_score,accuracy_score
        from sklearn.linear_model import LinearRegression,LogisticRegression
        data_type=self.data_type
        cat_col = [i for i in data_type.loc[data_type.loc[:,'Type']== 'categorical','Column'] if i != self.label_col]
        num_col = [i for i in data_type.loc[data_type.loc[:,'Type']== 'numerical','Column'] if i != self.label_col]
        S1,S2=[],[]
        for i in range(3):
            mini_batch = self.data.sample(n = min(20000,len(self.data))) 
            A = pd.DataFrame(SimpleImputer(strategy='mean').fit(mini_batch.loc[:,num_col]).transform(mini_batch.loc[:,num_col]), columns=num_col, index=mini_batch.index)
            # most frequent imputer
            if cat_col!=[]:
                B = mini_batch.loc[:,cat_col].fillna(pd.Series([mini_batch.loc[:,cat_col][c].value_counts().index[0] for c in mini_batch.loc[:,cat_col]],
                                                        index = mini_batch.loc[:,cat_col].columns))
                B = pd.get_dummies(B) # one hot encoding

                mini_batch = pd.concat([A,B,mini_batch.loc[:,self.label_col]],axis=1)
            else:
                mini_batch = pd.concat([A,mini_batch.loc[:,self.label_col]],axis=1)
                
            train_mini_batch = mini_batch.loc[mini_batch.index.isin(a.training_data),:]
            test_mini_batch =  mini_batch.loc[mini_batch.index.isin(a.testing_data),:]
            X_train = train_mini_batch.drop(columns=[self.label_col])
            Y_train = train_mini_batch.loc[:,self.label_col]
            X_test = test_mini_batch.drop(columns=[self.label_col])
            Y_test = test_mini_batch.loc[:,self.label_col]            
            encoder = LabelEncoder().fit(Y_train)
            
            Y_train = encoder.transform(Y_train)
            Y_test = encoder.transform(Y_test)
            
            clf = LogisticRegression()
            clf.fit(X_train, Y_train)
            Y_pred = clf.predict(X_test)
            y_score = clf.decision_function(X_test)

            S1.append(average_precision_score(Y_test, y_score))
            S2.append(accuracy_score(Y_test, Y_pred))
        print('Precison score: {}'.format(np.mean(S1)))
        print('classification accuracy: {}'.format(np.mean(S2)))
        warnings.filterwarnings("default")
        return None

    @property
    def col_name(self):
        return list(self.data) 
    
    @property
    def if_cleaned(self):
        return self.data_cleaned   
    
    @property
    def na_report(self):
        missing_df = self.data.isnull().sum(axis=0).reset_index()
        missing_df.columns = ['variable', 'missing values']
        missing_df['filling factor (%)']=(self.data.shape[0]-missing_df['missing values'])/self.data.shape[0]*100
        return missing_df.sort_values('filling factor (%)').reset_index(drop = True)
    
    @property
    def data_type(self):
        columns = list(self.data)
        converter = {'bool':'categorical','object':'categorical','int64':'numerical','float64':'numerical'}
        types = []
        samples = []
        unique_values = []
        for i in columns:
            type_ = str(self.data.loc[:,i].dtypes)
            if type_ == 'bool' or type_ == 'object':
                types.append(converter[type_])
                unique_values.append(len(self.data.loc[:,i].unique()))
            elif type_ == 'float64':
                types.append(converter[type_])
                unique_values.append(np.nan)
            elif len(self.data.loc[:,i].unique())<=10:
                types.append('categorical')
                unique_values.append(len(self.data.loc[:,i].unique()))
            else:
                types.append('numerical')
                unique_values.append(np.nan)
            samples.append(self.data.dropna().loc[0,i])
        return pd.DataFrame({'Column':columns,'Type':types,'Sample_text':samples,'unique_values':unique_values})        
    
    def noralization(self,columns=[],method='MaxMin'):
        data = self.data.loc[:,columns]
        training_data = data.loc[self.training_data,:]
        if method =='MaxMin':
            means = training_data.mean(axis = 0)
            # calculate max - min
            max_min = training_data.max(axis = 0) - training_data.min(axis = 0)
            # apply the transformation to data
            data = (data - means) / max_min
            
        elif method == 'Std':
            from sklearn.preprocessing import StandardScaler
            data.loc[:,:] = StandardScaler().fit(training_data).transform(data)
        
        elif method == 'Robust':
            from sklearn.preprocessing import RobustScaler
            data.loc[:,:] = RobustScaler().fit(training_data).transform(data)            
        elif method == 'MaxAbs':
            from sklearn.preprocessing import MaxAbsScaler
            data.loc[:,:] = preprocessing().fit(training_data).transform(data)   
        else:
            print('please select method from: Max_Min | Std | Robust | MaxAbs')
            raise        
        self.data.update(data) 
        self.update()
        return None
    
    def scale_transformation(self,cols = [],method = 'Yeo-Johnson',auto_para = 9):
        from sklearn.preprocessing import PowerTransformer
        # load your data
        data = self.data
        training_data = data.loc[self.training_data,:]
        if cols == []:
            skewness = training_data.skew()
            cols = [skewness.index[n] for n,i in enumerate(abs(skewness)>auto_para) if i == True and skewness.index[n] in self.num_col]
        # create the power transformer object with method 'box-cox'
        if method == 'Yeo-Johnson':
            data = pd.DataFrame(PowerTransformer(method='yeo-johnson', standardize=True).fit(training_data[cols]).transform(data[cols]), columns=cols, index=data.index)
        elif method == 'Box-Cox':
            data = pd.DataFrame(PowerTransformer(method='box-cox', standardize=True).fit(training_data[cols]).transform(data[cols]), columns=cols, index=data.index)
        self.data.update(data)
        return None     
    
    def outlier_removal(self,columns=[],limits=[0.05,0.95],method='Trim'):
        data = self.data.loc[:,columns]
        training_data = data.loc[self.training_data,:]
        if method == 'Trim':
            Q1 = training_data.quantile(limits[0])
            Q3 = training_data.quantile(limits[1])
            data = data[~((data < Q1) |(data > Q3)).any(axis=1)]
            self.data = self.data.loc[data.index,:]
            self.update()
        elif method == 'Inter-quantal':
            for variable in columns:
                IQR = training_data[variable].quantile(limits[1]) - training_data[variable].quantile(limits[0])
                lower = training_data[variable].quantile(limits[0]) - (IQR * 1.5)
                upper = training_data[variable].quantile(limits[1]) + (IQR * 1.5)
                data[variable] = np.where(data[variable] > upper, upper, np.where(data[variable] < lower, lower, data[variable]))
        elif method == 'Gaussian':
            for variable in columns:
                lower = training_data[variable].mean() - 3 * training_data[variable].std()
                upper = training_data[variable].mean() + 3 * training_data[variable].std()
                self.data[variable] = np.where(data[variable] > upper, upper, np.where(data[variable] < lower, lower, data[variable]))
        elif method == 'Quantiles':
            for variable in columns:
                lower = training_data[variable].quantile(limits[0])
                upper = training_data[variable].quantile(limits[1])
                self.data[variable] = np.where(data[variable] > upper, upper, np.where(data[variable] < lower, lower, data[variable]))
        else:
            print('please select method from: Trim | Inter-quantal | Gaussian | Quantiles')
            raise
        
        return None
    
    def discretization(self,columns,method='DecisionTree',n_bin=5,force_multi_iter = False):
        assert True not in [i in self.string_cat_col for i in columns] # make sure all numeric
        from sklearn.preprocessing import KBinsDiscretizer
        data = self.data.loc[:,columns+[self.label_col]]
        training_data = data.loc[self.training_data,:]
        if method == 'EqualWidth':
            data.loc[:,:] = KBinsDiscretizer(n_bins=n_bin, encode='ordinal', strategy='uniform').fit(training_data).transform(data)        
        
        elif method == 'EqualFrequency':
            data.loc[:,:] = KBinsDiscretizer(n_bins=n_bin, encode='ordinal', strategy='quantile').fit(training_data).transform(data)    
        
        elif method == 'Kmeans':
            data.loc[:,:] = KBinsDiscretizer(n_bins=n_bin, encode='ordinal', strategy='kmeans').fit(training_data).transform(data)    
            
        elif method == 'DecisionTree':
            assert self.label_col!=None
            from sklearn.tree import DecisionTreeClassifier
            
            target = training_data[self.label_col]
            for col in columns:
                x_variable = training_data.loc[:,[col]]
                # start the learning process.
                model = DecisionTreeClassifier(max_depth=2).fit(x_variable, target)
                data.loc[:,col] = model.predict_proba(data.loc[:,[col]])[:, 1]                
                a = pd.concat([self.data.loc[:,[col]],data.loc[:,[col]].rename(columns={col:'new'})],axis=1).sort_values(col)
                dict_ = {i:n for n,i in enumerate (a['new'].unique())}
                data.loc[:,col] = data.loc[:,col].apply(lambda x: dict_[x])
                
        elif method == 'MDLP_Entropy':
            # Minimum description length principle algorithm
            # https://github.com/maxpumperla/entropy-mdlp
            # Here i made addtional improvement to allow for additional iteration
            assert self.label_col!=None   
            # fit model only on training dataset
            y = training_data.loc[:,self.label_col]
            for col in columns:
                mdlp = MDLP()
                bins = list(mdlp.cut_points(training_data[col].values, y.values))
                values = mdlp.discretize_feature(training_data.reset_index().loc[:,col], bins)
                if force_multi_iter == True:
                    while len(bins)<n_bin-1:
                        for n,i in enumerate(set(values)):
                            X = training_data.loc[values==i,col]
                            Y = training_data.loc[values==i,self.label_col]
                            gain = -np.inf
                            cut_point = None
                            for i in training_data.loc[values==i,col].unique():
                                score = cut_point_information_gain_numpy(X, Y, i)
                                if score > gain:
                                    cut_point=i
                            bins.append(cut_point-0.00001)
                        bins.sort()
                # apply the bins on full feature
                values = mdlp.discretize_feature(data.reset_index().loc[:,col], bins)
                data[col] = values
                data[col] = np.where(data[col].isnull(), np.nan, data[col])  
                
        else:
            print('please select method from: EqualWidth | EqualFrequency | Kmeans | DecisionTree | MDLP_Entropy')
            raise  
        for col in columns:
            plt.figure(figsize=(7,3))
            plt.scatter(self.data[col],data[col])
            plt.title('feature: ' + col)
            plt.show()  
        data = data.astype('object')
        self.data.update(data) 
        self.update()
        return None     
    
    def get_top_abs_correlations(self, n=5,only_numeric=True):
        # select only numerical columns
        if only_numeric==True:
            df = self.data.select_dtypes(include='number')
        else:
            df = self.data
        
        au_corr = df.corr().abs().unstack()
        pairs_to_drop = set()
        cols = df.columns
        for i in range(0, df.shape[1]):
            for j in range(0, i+1):
                pairs_to_drop.add((cols[i], cols[j]))
    
        au_corr = au_corr.drop(labels=pairs_to_drop).sort_values(ascending=False)
        corr = df.corr().round(3)
        if len(df.columns)<=10:
            plt.figure(figsize=(9,9))
            sns.heatmap(corr, annot=True,center=0,vmin=-1, vmax=1,cmap='RdBu_r')
            plt.show()
        else:
            plt.figure(figsize=(9,9))        
            plt.imshow(corr,vmin=-1, vmax=1,cmap='RdBu_r')
            plt.colorbar()
            plt.show()
        return au_corr[0:n]
    
    def feature_encoder(self,columns,method='OneHot',target_dimension=1):
        df1 = self.data
        training_data = df1.loc[self.training_data,:]
        
        if method == 'OneHot':
            X = df1.loc[:,columns].astype('object')
            X = pd.get_dummies(X)
            df1 = pd.concat([df1.drop(columns=columns),X],axis=1)
            
        elif method =='TFIDF':
            # this need to use full data becasue it has no relation to our prediction column
            from my_utility.encoder import TfIdfEncoder
            assert len(columns)==1
            X = df1.loc[:,columns]
            ecoder = TfIdfEncoder(input_columns=['home_ownership'],tokens='words')
            ecoder.fit(X)
            X = ecoder.transform(X)
            X = pd.DataFrame(X.todense())
            X.columns = [ecoder.idx_to_token[i] for i in range(np.shape(X)[1])]
            df1 = pd.concat([df1.drop(columns=columns),X],axis=1)
            
        elif method=='EntityEmbedding':
            from tensorflow.keras import Input
            from tensorflow.keras.layers import Embedding, Reshape,Concatenate,Activation,Dense
            from tensorflow.keras.models import Model as KerasModel
            from sklearn.preprocessing import LabelEncoder,OneHotEncoder
            from tensorflow.keras.utils import to_categorical
            from typing import List, Tuple
            from sklearn.preprocessing import LabelBinarizer
            def label_encode(data: List) -> [np.ndarray, List[LabelEncoder]]:
                """
                This method is used to perform Label Encoding on a given list
                :param data: the list containing the items to be encoded
                :return: the encoded np.ndarray
                """
                labels_encoded = []
                data_encoded = np.array(data)
                for i in range(data_encoded.shape[1]):
                    le = LabelEncoder()
                    le.fit(data_encoded[:, i])
                    labels_encoded.append(le)
                    data_encoded[:, i] = le.transform(data_encoded[:, i])

                data_encoded = data_encoded.astype(int)
                return data_encoded, labels_encoded

            def sample(X: np.ndarray, y: np.ndarray, n: int) -> Tuple[np.ndarray, np.ndarray]:
                num_row = X.shape[0]
                indices = np.random.randint(num_row, size=n)
                return X[indices, :], y[indices]

            def transpose_to_list(X: np.ndarray) -> List[np.ndarray]:
                """
                :param X: the ndarray to be used as source
                :return: a list of nd.array containing the elements from the numpy array
                """

                features_list = []
                for index in range(X.shape[1]):
                    features_list.append(X[..., [index]])

                return features_list
            
            # model.fit section, using training data only
            X = training_data.loc[:,columns]
            Y = training_data.loc[:,self.label_col]
            encoded_X , labels=  label_encode(X)
            X = transpose_to_list(encoded_X)
            Y = to_categorical(Y)
            # Embedding layer for the 'Day of Week' feature

            if len(columns)>1:
                input_layers = []
                output_layers = []
                for col in columns:
                    n_unique_day = df1[col].nunique()
                    if target_dimension=='':
                        n_dim_day = int(np.sqrt(n_unique_day))
                    else:
                        n_dim_day = target_dimension
                    input_week = Input(shape=(1, ))
                    output_week = Embedding(input_dim=n_unique_day, 
                                            output_dim=n_dim_day, name=col)(input_week)
                    output_week = Reshape(target_shape=(n_dim_day, ))(output_week)
                    input_layers.append(input_week)
                    output_layers.append(output_week)
                model = Concatenate()(output_layers)
            else:
                col=columns[0]
                n_unique_day = df1[col].nunique()
                if target_dimension=='':
                    n_dim_day = int(np.sqrt(n_unique_day))
                else:
                    n_dim_day = target_dimension
                input_layers = Input(shape=(1, ))
                output_layers = Embedding(input_dim=n_unique_day, 
                                        output_dim=n_dim_day, name=col)(input_layers)
                output_layers = Reshape(target_shape=(n_dim_day, ))(output_layers)
                model = output_layers               
            model = Dense(200, kernel_initializer="uniform")(model)
            model = Activation('relu')(model)
            model = Dense(100, kernel_initializer="uniform")(model)
            model = Activation('relu')(model)
            model = Dense(len(Y[0]))(model)
            model = Activation('sigmoid')(model)
            model = KerasModel(inputs=input_layers, outputs=model)
            model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['mse'])
            model.fit(x=X,y=Y,epochs=10 , batch_size=128,verbose=0)
            
            
            # model.transform section, using the full data
            if target_dimension == 1:
                for n,name in enumerate(columns):
                    embeddings = model.get_layer(name).get_weights()[0]
                    types = labels[n].classes_
                    dict_ = {type_:embeddings[n] for n,type_ in enumerate(types)}
                    df1.loc[:,name] = df1.loc[:,name].apply(lambda x: dict_[x][0])
            else:
                for n,name in enumerate(columns):
                    embeddings = model.get_layer(name).get_weights()[0]
                    types = labels[n].classes_
                    dict_ = {type_:embeddings[n] for n,type_ in enumerate(types)}
                    for k in range(len(embeddings[0])):
                        df1.loc[:,name+'_'+str(k)] = df1.loc[:,name].apply(lambda x: dict_[x][k])
                df1 = df1.drop(columns = columns)    
        self.data = df1
        self.update()
        return None
    def merge_column(self,columns,target_col=['untitle']):
        '''
        columns = ['loan_amnt','grade','pub_rec','pub_rec_bankruptcies','addr_state']
        target_col = ['merge_name_1','merged_column_2']
        '''
        from sklearn.preprocessing import MinMaxScaler
        df = self.data
        training_data = df1.loc[self.training_data,:]
        
        X = training_data.loc[:,columns]
        X_all = df.loc[:,columns]
        
        N = len(target_col)
        categories = [i for i in columns if i in self.string_cat_col]
        numerics = [i for i in columns if i not in categories]

        if categories!=[]:
            X.loc[:,categories] = X.loc[:,categories].astype('object')
        if numerics!=[]:
            X.loc[:,numerics] = MinMaxScaler().fit_transform(X.loc[:,numerics])

        if numerics ==[]:
            mca = MCA(
                n_components=N,
                n_iter=3,
                copy=True,
                check_input=True,
                engine='auto',
                random_state=42
            )
            mca = mca.fit(X)
            new_df = mca.transform(X_all)
        elif categories ==[]:
            pca = PCA(n_components=N,
                    n_iter=10,
                    rescale_with_mean=True,
                    rescale_with_std=True,
                    copy=True,
                    check_input=True,
                    engine='auto',
                    random_state=42
            )
            pca = pca.fit(X)
            new_df = pca.transform(X_all)    
        else:
            famd = FAMD(
                n_components=N,
                n_iter=3,
                copy=True,
                check_input=True,
                engine='auto',
                random_state=42
            )
            famd = famd.fit(X)
            new_df = famd.transform(X_all)    

        new_df.columns = target_col
        df = df.drop(columns = columns)
        df = pd.concat([df,new_df],axis=1)
        self.data = df
        self.update()    
        return  self.data.head()         
    
    def imputate(self, method={},hyperparas = [128,0.5,100,10000]):
        '''
        This function aims to deal with missing datas
        hyperparas is only for GAIN method
        
        method is a dictionary {method:[columns]}, for example
        method = {
            'Drop' : [col1,col2]     # remove the rows that contain the missing value
            'Mean'   : [col]         # interpolate missing value using the mean/median value
            'MaxOccurance': [col]   # impute value based on most frequently occured category
            'KNN': [col,col,col...]  # looking at k most similar rows and impute way by the weighted average of rows
            'GAIN':[col]             # Generative Adversarial Imputation Networks (GAIN). Not recommend, too slow
            'SimpleDatawig': [col]  # Neural network based imputation. Not recommend, too slow
        }
        '''
        for i in method.keys():
            if i == 'Drop':
                self.data = self.data.dropna(subset=method[i])
                
            elif i == 'MaxOccurance':
                cat_col = method[i]
                self.data = self.data.loc[:,cat_col].fillna(pd.Series(
                                                                [self.data.loc[:,cat_col][c].value_counts().index[0] for c in self.data.loc[:,cat_col]],
                                                                index = self.data.loc[:,cat_col].columns))
            elif i == 'Mean':
                self.data.update(self.data.loc[:,method[i]].fillna(self.data.loc[:,method[i]].mean()))
                
            elif i == 'Median':
                self.data.update(self.data.loc[:,method[i]].fillna(self.data.loc[:,method[i]].median())) 
                
            elif i == 'Mode':
                df = self.data.loc[:,['emp_length_p','avg_cur_bal']].mode()
                self.data.update(self.data.loc[:,['emp_length_p','avg_cur_bal']].fillna({i:df.loc[0,i] for i in list(df)}))             
                
            elif i == 'MICE':
                import impyute
                data = self.data.loc[:,method[i]]
                data.loc[:,:] = impyute.imputation.cs.buck_iterative(data.values)
                self.data.update(data)
            
            elif i == 'KNN':
                import impyute
                df = self.data.loc[:,self.num_col]
                df.loc[:,:] = impyute.imputation.cs.fast_knn(df.values,k=5)
                self.data.update(df.loc[:,method[i]])
            elif i == 'GAIN':
                from GAIN import gain
                data = self.data
                data.loc[:,:] = gain.gain(data.values,{'batch_size':hyperparas[0],'hint_rate':hyperparas[1],'alpha':hyperparas[2],'iterations':hyperparas[3]})
                self.data.update(data.loc[:,method[i]])
                
            elif i == 'SimpleDatawig':
                self.data = datawig_simple(self.data,method[i])   
        self.update()
        return None
    
    def auto_feature_selection(self,n_features = 'auto'):
        from sklearn.feature_selection import SelectFromModel
        from sklearn.ensemble import RandomForestClassifier
        from sklearn.preprocessing import MinMaxScaler
        from sklearn.linear_model import LogisticRegression
        from lightgbm import LGBMClassifier
        from sklearn.feature_selection import SelectKBest
        from sklearn.feature_selection import chi2
        from sklearn.feature_selection import RFE
        from sklearn.preprocessing import LabelEncoder

        self.update()  # detect numerical and categorical features again
        if self.string_cat_col!=[]:
            self.feature_encoder(self.string_cat_col,method='EntityEmbedding')
        self.data.update(self.data.fillna(self.data.mean()))
        
        # feature selection based on training data onlu
        X = self.data.loc[self.training_data,self.num_col+self.cat_col]
        Y = self.data.loc[self.training_data,self.label_col]
        Y = LabelEncoder().fit(Y).transform(Y)
        X.loc[:,:] = MinMaxScaler().fit_transform(X)
        
        if n_features == 'auto':
            n_features = int(len(X.columns)/2)
        n_features_2 = min(n_features+2,len(X.columns))

        embeded_rf_selector = SelectFromModel(RandomForestClassifier(n_estimators=10), threshold=-np.inf, max_features=n_features_2)
        embeded_rf_selector.fit(X, Y)
        embeded_rf_support = embeded_rf_selector.get_support()
        embeded_rf_feature = X.loc[:,embeded_rf_support].columns.tolist()
        embeded_rf_feature


        embeded_lr_selector = SelectFromModel(LogisticRegression(penalty="l1",solver='liblinear'), threshold=-np.inf, max_features=n_features_2)

        embeded_lr_selector.fit(X, Y)
        embeded_lr_support = embeded_lr_selector.get_support()
        embeded_lr_feature = X.loc[:,embeded_lr_support].columns.tolist()



        lgbc=LGBMClassifier(n_estimators=10, learning_rate=0.05, num_leaves=32, colsample_bytree=0.2,
                    reg_alpha=3, reg_lambda=1, min_split_gain=0.01, min_child_weight=40)

        embeded_lgb_selector = SelectFromModel(lgbc, threshold=-np.inf, max_features=n_features_2)

        embeded_lgb_selector.fit(X, Y)
        embeded_lgb_support = embeded_lgb_selector.get_support()
        embeded_lgb_feature = X.loc[:,embeded_lgb_support].columns.tolist()




        chi_selector = SelectKBest(chi2, k=n_features_2)
        chi_selector.fit(X, Y)
        chi_support = chi_selector.get_support()
        chi_feature = X.loc[:,chi_support].columns.tolist()

        # recursive feature elimination
        rfe_selector = RFE(estimator=LogisticRegression(penalty='l1',solver='liblinear'), n_features_to_select=n_features_2, step=10, verbose=5)
        rfe_selector.fit(X, Y)
        rfe_support = rfe_selector.get_support()
        rfe_feature = X.loc[:,rfe_support].columns.tolist()




        feature_selection_df = pd.DataFrame({'Feature':X.columns, 'Chi-2':chi_support, 'RFE':rfe_support, 'Logistics':embeded_lr_support,
                                            'Random Forest':embeded_rf_support, 'LightGBM':embeded_lgb_support})

        feature_selection_df['Total'] = np.sum(feature_selection_df, axis=1)
        feature_selection_df = feature_selection_df.sort_values(['Total','Feature'] , ascending=False)
        feature_selection_df.index = range(1, len(feature_selection_df)+1)
        result = feature_selection_df.head(n_features)
        
        self.data = self.data.loc[:,[self.label_col]+list(result.Feature)]
        self.update()
        return result
    
    def auto_merge_high_correlation(self):
        '''
        The algorithm aims to find set of high correlation groups and merge each group into one feature
        '''
        corrMatrix=self.data.corr().abs()
        corrMatrix.loc[:,:] =  np.tril(corrMatrix, k=-1) # borrowed from Karl D's answer
        already_in = set()
        result = []
        for col in corrMatrix:
            perfect_corr = corrMatrix[col][corrMatrix[col] > 0.75].index.tolist()
            if perfect_corr and col not in already_in:
                already_in.update(set(perfect_corr))
                perfect_corr.append(col)
                result.append(perfect_corr)
        print(result)
        for columns in result:
            self.merge_column(columns,target_col=['+'.join(columns)])
        return None

    
class TimeSeries_DataFrame(DataFrame_Preprocessing):
    def __init__(self,df=[],label_col=None):
        if len(df)!=0:
            DataFrame_Preprocessing.__init__(self,df,True,label_col)
        else:
            rw, outlier_ind = random_walk_with_outliers(0, 1000, 0.01)
            rw2, outlier_ind2 = random_walk_with_outliers(0, 1000, 0.01)
            df = pd.DataFrame(data={'Series1':rw,'Series2':rw2})
            DataFrame_Preprocessing.__init__(self,df,True,label_col)
        cut = int(len(self.data)*2/3)
        self.training_data,self.testing_data = list(self.data.index[:cut]),list(self.data.index[cut:])
        self.update()
    
    def linear_interpolation(self,columns,method='FillBetween'):
        dataset = self.data.loc[:,columns]
        if method == 'FillBetween':
            for ii,i in enumerate(dataset.index):
                for jj,j in enumerate(dataset):
                    if str(dataset.iloc[ii,jj])=='nan':
                        try:
                            if str(dataset.iloc[ii+1,jj])=='nan':
                                dataset.iloc[ii,jj]=dataset.iloc[ii-1,jj]
                            else:
                                dataset.iloc[ii,jj]=(dataset.iloc[ii-1,jj]+dataset.iloc[ii+1,jj])/2
                        except:
                                dataset.iloc[ii,jj]=dataset.iloc[ii-1,jj]
                
        elif method == 'EMW':
            # exponetially weighted average
            import impyute
            dataset.loc[:,:]=impyute.imputation.ts.moving_window(dataset,wsize=5)
            
        else:
            print('please select method from: FillBetween | EMW')
            raise              
        self.data.update(dataset)
        return None
        
    def hampel_filter(self,columns, window_size=10, n_sigmas=3):
        # new hamper filter
        # https://towardsdatascience.com/outlier-detection-with-hampel-filter-85ddf523c73d
        for col in columns:
            new_series = self.data.loc[:,col]
            n = len(new_series)
            k = 1.4826 # scale factor for Gaussian distribution
            indices = []
            # possibly use np.nanmedian 
            for i in range((window_size),(n - window_size)):
                x0 = np.median(new_series[(i - window_size):(i + window_size)])
                S0 = k * np.median(np.abs(new_series[(i - window_size):(i + window_size)] - x0))
                if (np.abs(new_series[i] - x0) > n_sigmas * S0):
                    new_series[i] = x0
                    indices.append(i)
            self.data.loc[:,col] = new_series   
        return None
    
    def get_change_rate(self,columns):
        return (self.data.loc[:,columns]-self.data.loc[:,columns].shift(1))/self.data.loc[:,columns].shift(1)
    
    
class Non_TimeSeries_DataFrame(DataFrame_Preprocessing):
    def __init__(self,df,label_col=None):
        from sklearn.model_selection import train_test_split
        DataFrame_Preprocessing.__init__(self,df,True,label_col)
        a,b = train_test_split(self.data,test_size=0.33, random_state=42)
        self.training_data,self.testing_data=list(a.index),list(b.index)
        self.update()

    
```

```python
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE


training_data = a.data.loc[a.training_data,:]
training_data = pd.concat([pd.get_dummies(training_data.loc[:,a.cat_col].astype('object')),training_data.loc[:,a.num_col+[a.label_col]]],axis=1)



X = training_data.drop(a.label_col, axis=1)
Y = training_data.loc[:,a.label_col]
Y = LabelEncoder().fit(Y).transform(Y)

# split training dataset into training and validation dataset
X_train, X_vali, y_train, y_vali = train_test_split(X, Y,
                                                    stratify=Y,
                                                    test_size=0.20,
                                                    random_state=30)

def recall_m(y_true, y_pred):
    true_positives = keras.backend.sum(keras.backend.round(keras.backend.clip(y_true * y_pred, 0, 1)))
    possible_positives = keras.backend.sum(keras.backend.round(keras.backend.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + keras.backend.epsilon())
    return recall

def precision_m(y_true, y_pred):
    true_positives = keras.backend.sum(keras.backend.round(keras.backend.clip(y_true * y_pred, 0, 1)))
    predicted_positives = keras.backend.sum(keras.backend.round(keras.backend.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + keras.backend.epsilon())
    return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+keras.backend.epsilon()))
f1_m.__name__ = 'f1_score'

METRICS = [
      keras.metrics.Precision(name='precision'),
      keras.metrics.Recall(name='recall'),
      keras.metrics.AUC(name='auc'),
      f1_m
]


sm = SMOTE(sampling_strategy='auto', k_neighbors=10, random_state=10)
X_train, y_train = sm.fit_resample(X_train, y_train)
print(sorted(Counter(y_train).items()))
```

```python
class MyHyperModel(HyperModel):
    def __init__(self, num_classes):
        self.num_classes = num_classes
    def build(self, hp):
        # Specify model
        model = keras.Sequential()
        # Range of models to build
        for i in range(hp.Int('num_layers', 2, 10)):

            model.add(keras.layers.Dense(units=hp.Int('units_' + str(i),
                                                min_value=64,
                                                max_value=512, 
                                                step=32),
                                   activation='relu'))
            model.add(
                keras.layers.Dropout(hp.Float('dropout',min_value=0.0,max_value=0.1,default=0.005,step=0.01))
            )        
        # Output layer
        model.add(keras.layers.Dense(self.num_classes, activation='sigmoid'))
        # Compile the constructed model and return it
        model.compile(
            optimizer=keras.optimizers.Adam(
                hp.Choice('learning_rate',
                          values=[1e-2, 1e-3, 1e-4])),
            loss='binary_crossentropy',
            metrics=METRICS)
        return model

hypermodel = MyHyperModel(num_classes=1)

tuner = Hyperband(
    hypermodel,
    objective=kerastuner.Objective("f1_score", direction="max"),
    max_epochs=10,
    seed=10,
    overwrite=True)

# Search for the best parameters of the neural network using the contructed Hypberband tuner
tuner.search(X_train.values, y_train,
             epochs=10,
             validation_data=(X_vali.values, y_vali))

# Get the best hyperparameters from the search
params = tuner.get_best_hyperparameters()[0]

# Build the model using the best hyperparameters
model = tuner.hypermodel.build(params)

# Train the best fitting model
model.fit(X.values, Y, epochs=20)

# Check the accuracy plots
hyperband_accuracy_df = pd.DataFrame(model.history.history)

hyperband_accuracy_df[['loss', 'auc','f1_score']].plot()
plt.title('Loss & Accuracy Per EPOCH')
plt.xlabel('EPOCH')
plt.ylabel('Accruacy')
plt.show()
```

```python
from tensorflow.keras import Input
from tensorflow.keras.layers import Embedding, Reshape,Concatenate,Activation,Dense
from tensorflow.keras.models import Model as KerasModel
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from tensorflow.keras.utils import to_categorical
from typing import List, Tuple
from sklearn.preprocessing import LabelBinarizer
def label_encode(data: List) -> [np.ndarray, List[LabelEncoder]]:
    """
    This method is used to perform Label Encoding on a given list
    :param data: the list containing the items to be encoded
    :return: the encoded np.ndarray
    """
    labels_encoded = []
    data_encoded = np.array(data)
    for i in range(data_encoded.shape[1]):
        le = LabelEncoder()
        le.fit(data_encoded[:, i])
        labels_encoded.append(le)
        data_encoded[:, i] = le.transform(data_encoded[:, i])

    data_encoded = data_encoded.astype(int)
    return data_encoded, labels_encoded

def sample(X: np.ndarray, y: np.ndarray, n: int) -> Tuple[np.ndarray, np.ndarray]:
    num_row = X.shape[0]
    indices = np.random.randint(num_row, size=n)
    return X[indices, :], y[indices]

def transpose_to_list(X: np.ndarray) -> List[np.ndarray]:
    """
    :param X: the ndarray to be used as source
    :return: a list of nd.array containing the elements from the numpy array
    """

    features_list = []
    for index in range(X.shape[1]):
        features_list.append(X[..., [index]])

    return features_list

target_dimension = 1
df1 = pd.read_csv('credit.csv')
a = Non_TimeSeries_DataFrame(df = df1.drop(columns = ['Unnamed: 0','issue_d']),label_col='def_flag')
df1 =a.data
columns = ['grade']
X = df1.loc[:,columns]
Y = df1.loc[:,'def_flag']
encoded_X , labels=  label_encode(X)
X = transpose_to_list(encoded_X)


Y =to_categorical(Y)
# Embedding layer for the 'Day of Week' feature

input_layers = []
output_layers = []
for col in columns:
    n_unique_day = df1[col].nunique()
    if target_dimension=='':
        n_dim_day = int(np.sqrt(n_unique_day))
    else:
        n_dim_day = target_dimension
    input_week = Input(shape=(1, ))
    output_week = Embedding(input_dim=n_unique_day, 
                            output_dim=n_dim_day, name=col)(input_week)
    output_week = Reshape(target_shape=(n_dim_day, ))(output_week)
    input_layers.append(input_week)
    output_layers.append(output_week)


model = output_week
model = Dense(200, kernel_initializer="uniform")(model)
model = Activation('relu')(model)
model = Dense(100, kernel_initializer="uniform")(model)
model = Activation('relu')(model)
model = Dense(len(Y[0]))(model)
model = Activation('sigmoid')(model)
model = KerasModel(inputs=input_week, outputs=model)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['mse'])
model.fit(x=X,y=Y,epochs=10 , batch_size=128,verbose=0)

if target_dimension == 1:
    for n,name in enumerate(columns):
        embeddings = model.get_layer(name).get_weights()[0]
        types = labels[n].classes_
        dict_ = {type_:embeddings[n] for n,type_ in enumerate(types)}
        df1.loc[:,name] = df1.loc[:,name].apply(lambda x: dict_[x][0])
    df2 = df1
else:
    for n,name in enumerate(columns):
        embeddings = model.get_layer(name).get_weights()[0]
        types = labels[n].classes_
        dict_ = {type_:embeddings[n] for n,type_ in enumerate(types)}
        for k in range(len(embeddings[0])):
            df1.loc[:,name+'_'+str(k)] = df1.loc[:,name].apply(lambda x: dict_[x][k])
    df2 = df1.drop(columns = columns)    
```

**showcase of discretization**

### Equal-Width Discretization
- Width = (maximum value - minimum value) / N
- Equal Width doesnt improve the value spread
- It can handle outliers
- Can be combined with categorical encodings

### Equal-Frequency Discretization
- Equal Frequency does improve the value spread
- It can handle outliers
- Can be combined with categorical encoding

### K-Means Discretization
- K-Means doesnt improve the value spread
- It can handle outliers, however a centroid bias may exist.
- Can be combined with categorical encoding

### Discretization with Decision Trees
- Decision Tree does not improve the value spread
- It can handle outliers well as trees are robust to outliers.
- Creates monotonic relationships
